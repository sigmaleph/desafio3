```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('train/train_data.csv')
```

```python
df.shape
```

```python
for col in df.columns:
    if col!='index':
        print(col)
        print(df[col].value_counts())
```

```python
coldata=[]

for col in df.columns:
    tipo=type(df.loc[0,col])
    if tipo==type(''):
        if df.loc[0,col][0]=='[':
            tipo=type([])
    dist=len(df[col].value_counts())
    nulls=df[col].isnull().sum()
    coldata.append({'column_name':col, 'type':tipo, 'distinct_values':dist, 'null_values':nulls})
```

```python
infodf=pd.DataFrame(coldata)
```
